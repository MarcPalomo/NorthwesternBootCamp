## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# group by SN
screen_name = purchase_data.groupby(['SN'])

# get the values for the grouped data
sn_purchase_count = screen_name["Price"].count()
sn_avg_purchase = screen_name["Price"].mean()
sn_total_purchase = screen_name["Price"].sum()
```

```python
# Creating a new DataFrame using above values
sn_summary_table = pd.DataFrame({"Purchase Count": sn_purchase_count, 
                                 "Average Purchase Price": sn_avg_purchase, 
                                 "Total Purchase Value": sn_total_purchase})

# sort the values by Total Purchase Value
sn_sorted = sn_summary_table.sort_values(by="Total Purchase Value", ascending=False)

# clean formatting
sn_sorted["Average Purchase Price"] = sn_sorted["Average Purchase Price"].astype(float).map("${:,.2f}".format)
sn_sorted["Total Purchase Value"] = sn_sorted["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# display
sn_sorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# retrieve Item ID, Item Name, Item Price
purchase_data_new = purchase_data[["Item ID", "Item Name", "Price"]].copy()

# group data by Item ID and Item Name
popular = purchase_data_new.groupby(["Item ID", "Item Name"])

# get the values for the grouped data
popular_count = popular["Price"].count()
popular_avg_purchase = popular["Price"].mean()
popular_total_purchase = popular["Price"].sum()
```

```python
# Creating a new DataFrame using above values
popular_summary_table = pd.DataFrame({"Purchase Count": popular_count, 
                                 "Item Price": popular_avg_purchase, 
                                 "Total Purchase Value": popular_total_purchase})

# sort the values by Total Purchase Value
popular_sorted = popular_summary_table.sort_values(by="Purchase Count", ascending=False)

# clean formatting
popular_sorted["Item Price"] = popular_sorted["Item Price"].astype(float).map("${:,.2f}".format)
popular_sorted["Total Purchase Value"] = popular_sorted["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# display
popular_sorted.head()
```
