```python
frames = [BD_07, BD_08, BD_09, BD_10, BD_11, BD_12, BD_01, BD_02, BD_03, BD_04, BD_05, BD_06]
```

```python
new = result["starttime"].str.split(" ", n=1, expand = True)
result["Start_Date"] = new[0]
```

```python
result["Start_Time"] = new[1]
```

```python
new2 = result["stoptime"].str.split(" ", n=1, expand = True)
result["Stop_Date"] = new2[0]
```

```python
result["Stop_Time"] = new2[1]
```

```python
result.drop(columns=["starttime"], inplace = True)
```

```python
result.drop(columns=["stoptime"], inplace = True)
```

```python
result.head()
```

```python
result.dtypes
```

```python
time = result["Start_Time"].str.split(".", n=1, expand = True)
result["Start"] = time[0]
```

```python
time2 = result["Stop_Time"].str.split(".", n=1, expand = True)
result["Stop"] = time2[0]
```

```python
result.drop(columns=["Start_Time"], inplace = True)
```

```python
result.drop(columns=["Stop_Time"], inplace = True)
```

```python
result.head()
```

```python
result.dtypes
```

```python
result.head()
```

```python
breakdown = result["Start"].str.split(":", n=2, expand = True)
result["Start_H"] = breakdown[0]
result["Start_M"] = breakdown[1]
result["Start_S"] = breakdown[2]
result.drop(columns=["Start"], inplace = True)
```

```python
breakdown2 = result["Stop"].str.split(":", n=2, expand = True)
result["Stop_H"] = breakdown2[0]
result["Stop_M"] = breakdown2[1]
result["Stop_S"] = breakdown2[2]
result.drop(columns=["Stop"], inplace = True)
```

```python
result.head()
```

```python
result.to_csv("Bike_Data.csv")
```
